```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# 旅游网站
- 提取酒店评分
- 提取酒店等级

```python
data = pd.read_csv(open('旅游.csv',encoding='utf-8'),index_col=0)
data.head()
```

```python
# 提取酒店评分
data['酒店评分'] = data['酒店'].str.extract(' (\d\.\d)分/5分')
data.head()
```

```python
# 提取酒店等级
data['酒店等级'] = data['酒店'].str.extract(' (.*) ')
data.head()
```

```python
data['天数']=data['路线名'].str.extract('(\d)天\d晚')
data.head()
```

# 租房数据提取

```python
zufang = pd.read_csv(open('租房.csv',encoding='utf-8'))
zufang.head()
```

```python
# 使用正则表达式提取网址
zufang['url'] = zufang['house_title'].str.extract('(https://.*)')
zufang.head()
```

```python
# 拼接几室几厅几卫
def str_concat(a,b,c):
    return str(a)+ '室' + str(b) + '厅' + str(c) + '卫'
zufang['户型'] = zufang.apply(lambda x : str_concat(x['bedroom_num'],x['hall_num'],x['bathroom_num']), axis=1)
zufang.head()
```

# 评论数据项目实战

```python
pinglun = pd.read_csv(open('评论.csv',encoding='utf-8'))
pinglun.head()
```

```python
## 用户活跃时间段分析
time = pinglun['creationTime'].str.extract(' (.*)')
time.head()
```

```python
pinglun['hour'] = time.str.extract('^(\d{2}):')
pinglun.head()
```

# 柱状图绘制，关注哪一个时间段更活跃

```python
a = np.sort(pinglun['hour'].unique())
a
```

```python
height = pinglun.groupby('hour').count()['content']
height
```

```python
font = {'family':'SimHei','size':15}
plt.rc('font',**font)

plt.figure(figsize=(20,5))

plt.bar(a,height=height)
plt.title('用户评论活跃时间段')
gca = plt.gca()
gca.spines['right'].set_color('none')
gca.spines['top'].set_color('none')

plt.show()
```
