```python
import pymysql
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
db_info={
    "host":"localhost", # 我们要连接的数据库的IP地址或者是域名，如果是本地的话可以写localhost或者是127.0.0.1
    "user":"root", # 我们连接数据库的用户名
    "password":"root", # 我们连接数据库的密码
    "database":"management_systems", # 我们要连接的数据库的名字
    "charset":"utf8" # 这是字符集
}
conn = pymysql.connect(**db_info)# 开始连接
cursor = conn.cursor()# cursor是执行sql语句的对象
```

```python
sql = 'select * from customer'
data = pd.read_sql(sql,conn)
data.head()
```

```python
# 0行多余 重设索引
data = data.set_index('id')
data.head()
```

```python
## 修改列名
data.columns = ['用户id','订单时间','订单数量','订单金额']
data.head()
```

# 复购和回购

- 统计用户消费次数

```python
pivoted_counts = data.pivot_table(index='用户id',columns='月份',values='订单时间',aggfunc='count').fillna(0)#pivot_table数据透视表
pivoted_counts
```

# 确定用户是否有回购的行为

```python
pivoted_counts.transf = pivoted_counts.applymap(lambda x: 1 if x >1 else np.NaN if x==0 else 0)
pivoted_counts.transf.head()
```

# 每个月的复购率

```python
plt.figure(figsize=(20,5))
month_counts_reorder_rate = pivoted_counts.transf.sum()/pivoted_counts.transf.count()#回购总数/回购数
plt.plot(month_counts_reorder_rate)
plt.title('每月的复购率')
plt.xlabel('月份')
plt.ylabel('百分比')
plt.show()
```

# 每月消费和二次消费的用户数

```python
a,b = plt.subplots(figsize=(20,5))
b.plot(pivoted_counts.transf.count())
b.plot(pivoted_counts.transf.sum())
b.legend(['消费人数','二次消费以上的人数'])
plt.title('每月消费和二次消费的用户数')
plt.xlabel('月份')
plt.ylabel('用户数')
plt.show()
```
