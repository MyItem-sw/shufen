# 学习一下时间戳
- 时间戳是是从1970年1月1日早上8点开始，到现在某一时刻的毫秒值

```python
data['订单开始时间']
```

# 把时间转换为时间戳

```python
dtime = data['订单开始时间']

v = (dtime.values - np.datetime64('1970-01-01 08:00:00'))/np.timedelta64(1,'ms')
v
```

```python
data['订单开始时间']= v
data.head()
```

```python
data.info()
```

# 时间戳转换为时间

```python
data['订单开始时间'] = pd.to_datetime(data['订单开始时间'],unit='ms',origin=pd.Timestamp('1970-01-01 08:00:00'))
data.head()
```
